```python
import numpy as np
import matplotlib.pyplot as plt
from fractions import Fraction
import CT_classical

# Assuming CT_classical has already been defined as per the code provided

# Parameters
pm = 0.5  # Probability to apply control map
num_iterations = 1000

ct_classical = CT_classical.CT_classical(L=10, x0=1/np.sqrt(2), xj=[Fraction(1, 3), Fraction(2, 3)], store_vec=True)

trajectory = []
for _ in range(num_iterations):
    ct_classical.random_control(pm)

for integer in ct_classical.vec_history:
    # convert the integer state vector to decimal
    trajectory.append(integer / (1 << ct_classical.L))

# Plot the trajectory
plt.figure(figsize=(10, 6))
plt.plot([integer_state / (1 << ct_classical.L) for integer_state in ct_classical.vec_history], 'o', markersize=4, label='x Trajectory')
plt.axhline(y=1/3, color='r', linestyle='--', label='1/3')
plt.axhline(y=2/3, color='g', linestyle='--', label='2/3')
plt.xlabel('Iteration')
plt.ylabel('x value')
plt.title('Trajectory of x under Probabilistic Control Map and Bernoulli Map')
plt.legend()
plt.grid(True)
plt.show()

# Plot Neel order parameter
plt.figure(figsize=(10, 6))
plt.plot([ct_classical.ZZ(integer_state) for integer_state in ct_classical.vec_history], 'o', markersize=4, label='ZZ Neel order')
plt.xlabel('Iteration')
plt.ylabel('ZZ')
plt.title('Neel order of x under Probabilistic Control Map and Bernoulli Map')
plt.legend()
plt.grid(True)
plt.show()
```
